# Requirements
To run this notebook install the following:

* pip install earthkit-plots
* pip install cf-units         (Optional for unit conversion in maps)

```python
import earthkit.data
import earthkit.regrid


request = {
    "dataset": "extremes-dt",
    "class": "d1",
    "stream": "oper",
    "type": "fc",
    "date": "20250119",
    "time": "0000",
    "levtype": "pl",
    "expver": "0001",
    "param": "130",
    "feature" :{ 
        "type" : "trajectory", 
        "points" : [[-1, -1, 1, 1], [0, 0, 2, 5], [1, 1, 10, 100]], 
        "inflation" : 0.1, 
        "inflate" : "box",
        "axes" : ["latitude", "longitude", "step", "levelist"], 
    },
}

#data is an earthkit streaming object but with stream=False will download data immediately 
ds = earthkit.data.from_source("polytope", "destination-earth", request, stream=False, address='polytope-test.lumi.apps.dte.destination-earth.eu')
```

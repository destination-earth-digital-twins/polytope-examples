```python
import earthkit.data
import earthkit.plots
import earthkit.regrid
from polytope.api import Client
```

```python
# Set True if you want to make a live request for the data, or false if you want to use the cached grib file
LIVE_REQUEST = True
```

The default interpolation of healpix to healpix grids is ring nested, passing `'grid':'h16'` in your request will result in a healpix ring ordered grid to be returned. One can explicitly asks for nested by adding `n` to the gridspec `'grid' : 'hn16'`.

```python
request = {
    'activity': 'scenariomip',
    'class': 'd1',
    'dataset': 'climate-dt',
    'date': '20300101',
    'time': "0000",
    'experiment': 'SSP3-7.0',
    'expver': '0001',
    'generation': '1',
    'levtype': 'sfc',
    'model': 'IFS-NEMO',
    'param': '167',
    'realization': '1',
    'resolution': 'high',
    'stream': 'clte',
    'grid':'h16', # currently O, F, N, H grids are supported 
    'type': 'fc',
}
```

```python
data_file = "data/climate-dt-earthkit-healpix-interpolate.grib"
if LIVE_REQUEST:
    data = earthkit.data.from_source("polytope", "destination-earth", request, address="polytope.lumi.apps.dte.destination-earth.eu", stream=False)
    data.to_target("file", data_file)
else:
    data = earthkit.data.from_source("file", data_file) 
```

```python
data.ls()
```

```python
data.to_xarray()
```

```python
chart = earthkit.plots.Map(extent=[-180, 180, -90, 90])
chart.grid_cells(
    data,
    auto_style=True,
    units="celsius"
)

chart.title("Temperature")
chart.coastlines()
chart.gridlines()
chart.legend()
chart.show()
```

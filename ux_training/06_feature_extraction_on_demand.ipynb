## Select Your Country and Prepare Polygon

**EXERCISE**: Choose your country. Note the longitude remapping required for the On-Demand DT.

```python
# EXERCISE: Replace with your country
countries = ["Ireland"]  # <- YOUR COUNTRY HERE

shapes = cartography.country_polygons(countries, resolution=500e5)

# Remap longitudes from [-180, 180] to [0, 360] (required for On-Demand DT)
for shape in shapes:
    for point in shape:
        lon = point[1]
        if lon < 0:
            point[1] = lon + 360

print(f"Prepared {len(shapes)} polygon(s) for {countries}")
```

## Build the Request

**EXERCISE**: You also need to specify the correct `georef` geohash that covers your country. Look up your country's approximate center at [geohash.org](http://geohash.org).

```python
request = {
    "class": "d1",
    "dataset": "on-demand-extremes-dt",
    "expver": "0099",
    "stream": "oper",
    "date": "20250616",
    "time": 0,
    "type": "fc",
    "levtype": "sfc",
    "georef": "gc7x9",  # EXERCISE: Change to a geohash covering your country
    "step": 12,
    "param": 167,
    "feature": {
        "type": "polygon",
        "shape": shapes,
    },
}
```

## Retrieve the Data

```python
if LIVE_REQUEST:
    data = earthkit.data.from_source(
        "polytope", "destination-earth", request,
        address="polytope.lumi.apps.dte.destination-earth.eu",
        stream=False,
    )
else:
    data = earthkit.data.from_source("file", "../on-demand-extremes-dt/data/on-demand-fe-country-training.grib")
```

## Build the Request

The On-Demand Extremes DT uses a `georef` parameter (a geohash) to specify the geographic domain.

The geohash `u4usq2` covers part of central Europe.

```python
request = {
    "class": "d1",
    "dataset": "on-demand-extremes-dt",
    "expver": "0099",
    "stream": "oper",
    "date": "20250616",
    "time": 0,
    "type": "fc",
    "levtype": "sfc",
    "georef": "u4usq2",  # EXERCISE: Replace with a geohash for your region
    "step": 12,
    "param": 167,
}
```

## Retrieve the Data

```python
data_file = "../on-demand-extremes-dt/data/on-demand-basic-training.grib"

if LIVE_REQUEST:
    data = earthkit.data.from_source(
        "polytope", "destination-earth", request,
        address="polytope.lumi.apps.dte.destination-earth.eu",
        stream=False,
    )
    data.to_target("file", data_file)
else:
    data = earthkit.data.from_source("file", data_file)
```

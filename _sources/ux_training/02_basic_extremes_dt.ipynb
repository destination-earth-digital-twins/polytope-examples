## Build the Request

The Extremes DT uses relative dates (e.g. `-14` means 14 days ago) and forecast steps.

**EXERCISE**: Change the `step` to retrieve hour 6 of the forecast instead of hour 0. Also change `param` from `167` (2m temperature) to `228228` (total precipitation).

```python
request = {
    "class": "d1",
    "dataset": "extremes-dt",
    "expver": "0001",
    "stream": "oper",
    "date": "-14",  # 14 days ago
    "time": "0000",
    "type": "fc",
    "levtype": "sfc",
    "step": "0",  # EXERCISE: Change to step 6
    "param": "167",  # EXERCISE: Change to 228228 (total precipitation)
}
```

## Retrieve the Data

```python
data_file = "../extremes-dt/data/extremes-dt-basic-training.grib"

if LIVE_REQUEST:
    data = earthkit.data.from_source(
        "polytope", "destination-earth", request,
        address="polytope.lumi.apps.dte.destination-earth.eu",
        stream=False,
    )
    data.to_target("file", data_file)
else:
    data = earthkit.data.from_source("file", data_file)
```

## Inspect the Data

```python
data.ls()
```

## Plot the Data

```python
chart = earthkit.plots.Map(extent=[-180, 180, -90, 90])
chart.contourf(data[0], auto_style=True)
chart.coastlines()
chart.gridlines()
chart.title("Extremes DT - 2m Temperature")
chart.show()
```

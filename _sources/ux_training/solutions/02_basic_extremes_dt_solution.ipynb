## Imports

```python
import earthkit.data
import earthkit.plots
import os
```

## Live Request Toggle

```python
LIVE_REQUEST = os.getenv("LIVE_REQUEST", "true").lower() == "true"
LIVE_REQUEST
```

## Build the Request

The Extremes DT uses relative dates (e.g. `-14` means 14 days ago) and forecast steps.

```python
request = {
    "class": "d1",
    "dataset": "extremes-dt",
    "expver": "0001",
    "stream": "oper",
    "date": "-14",
    "time": "0000",
    "type": "fc",
    "levtype": "sfc",
    "step": "6",
    "param": "228228",
}
```

## Retrieve the Data

```python
data_file = "../../extremes-dt/data/extremes-dt-basic-training.grib"

if LIVE_REQUEST:
    data = earthkit.data.from_source(
        "polytope", "destination-earth", request,
        address="polytope.lumi.apps.dte.destination-earth.eu",
        stream=False,
    )
    data.to_target("file", data_file)
else:
    data = earthkit.data.from_source("file", data_file)
```

## Inspect the Data

```python
data.ls()
```

## Plot the Data

```python
chart = earthkit.plots.Map(extent=[-180, 180, -90, 90])
chart.grid_cells(data[0], style="auto")
chart.coastlines()
chart.gridlines()
chart.title("Extremes DT - Total Precipitation (Step 6)")
chart.show()
```

## Subset to Europe

```python
chart = earthkit.plots.Map(domain="europe")
chart.grid_cells(data[0], style="auto")
chart.coastlines()
chart.gridlines()
chart.title("Extremes DT - Total Precipitation (Europe)")
chart.show()
```
